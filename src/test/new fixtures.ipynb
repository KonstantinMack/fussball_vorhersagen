```python
import pandas as pd
```

```python
URL = "http://www.football-data.co.uk/mmz4281/1819/"
```

```python
def get_data():
    eng = pd.read_csv(URL + "E0.csv")
    fra = pd.read_csv(URL + "F1.csv")
    ger = pd.read_csv(URL + "D1.csv")
    ita = pd.read_csv(URL + "I1.csv")
    spa = pd.read_csv(URL + "SP1.csv")
    fixtures = pd.read_csv("http://www.football-data.co.uk/fixtures.csv")
    return eng, fra, ger, ita, spa, fixtures
```

```python
def get_fixtures(fixtures):
    cols = ['Div', 'Date', 'HomeTeam', 'AwayTeam', 'BbMxH', 'BbAvH', 'BbMxD', 'BbAvD',
           'BbMxA', 'BbAvA', 'BbOU', 'BbMx>2.5', 'BbAv>2.5', 'BbMx<2.5',
           'BbAv<2.5', 'BbAH', 'BbAHh', 'BbMxAHH', 'BbAvAHH', 'BbMxAHA',
           'BbAvAHA']
    eng_fixs = fixtures[fixtures["Div"] == "E0"][cols]
    fra_fixs = fixtures[fixtures["Div"] == "F1"][cols]
    ger_fixs = fixtures[fixtures["Div"] == "D1"][cols]
    ita_fixs = fixtures[fixtures["Div"] == "I1"][cols]
    spa_fixs = fixtures[fixtures["Div"] == "SP1"][cols]
    return eng_fixs, fra_fixs, ger_fixs, ita_fixs, spa_fixs
```

```python
def get_new_matches():
    eng, fra, ger, ita, spa, fixtures = get_data()
    eng_fixs, fra_fixs, ger_fixs, ita_fixs, spa_fixs = get_fixtures(fixtures)
    eng = pd.concat([eng, eng_fixs]).reset_index(drop=True)
    fra = pd.concat([fra, fra_fixs]).reset_index(drop=True)
    ger = pd.concat([ger, ger_fixs]).reset_index(drop=True)
    ita = pd.concat([ita, ita_fixs]).reset_index(drop=True)
    spa = pd.concat([spa, spa_fixs]).reset_index(drop=True)
    return eng, fra, ger, ita, spa, fixtures
```

```python
eng, fra, ger, ita, spa, fixtures = get_new_matches()
```
